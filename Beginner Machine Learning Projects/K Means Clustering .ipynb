Now this dataset is cleaned

Now I need to implement the algorithm, by creating cluster labels, and implementing the K means model with 2 clusters. 

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

Then I fit the data to all the data except the Private column

```python
kmeans.fit(df.drop('Private', axis = 1))
```

## Evaluation of the Algorithm

First I created a function to convert the Private column to 0 and 1 values, which we can interpret more easily. 



```python
def converter(x):
    if x == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```

Not a bad accuracy number!
